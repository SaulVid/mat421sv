Ordinary Differential Equations in python

```python
import numpy as np
import scipy as sp
import math
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy.integrate import solve_ivp

def dydx(y, x):
    return y + x**2
y0 = 0

t = np.linspace(0, 1, 100)
ans1 = odeint(dydx, y0=y0, t=t, tfirst=True)


ans2 = solve_ivp(dydx, t_span=(0,max(t)), y0=[y0], t_eval=t)


solution1 = ans1.T[0]

solution2 = ans2.y[0]

plt.plot(t, solution1)
plt.plot(t, solution2, '--')
plt.ylabel('$y(x)$', fontsize=10)
plt.xlabel('$x$', fontsize=10)
plt.show()
```

Euler's Method in Python


```python
import math

dy = lambda x,y: y
f = lambda x: math.e**x

xi = 0; xf=1; h=0.1
n=int((xf-xi)/h)
x=0; y=1

print ('x \t\t y'); print('%f \t %f'%(x,y))
x_plot = []; y_euler = []; y_analytical= []


for i in range (1,n+1):
  x_plot.append(x); y_euler.append(y); y_analytical.append(f(x))
  y = y+dy(x,y) * h
  x = x + h 
  print('%f \t %f'%(x,y))

x_plot.append(x); y_euler.append(y); y_analytical.append(f(x))
plt.plot(x_plot, y_euler, x_plot, y_analytical)
plt.legend(['Euler method','Analytical'])
```
